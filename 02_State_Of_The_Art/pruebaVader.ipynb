```python
### Preparación de la librería de insultos
# Obtenida de https://listas.20minutos.es/lista/insultos-en-castellano-que-deberias-conocer-y-su-significado-393340/
path_lib_insultos="C:/Users/Esperanza/Desktop/UOC/TFM/python/libreria_insultos.txt"

# necesitamos quitar los acentos y las diéresis para no sustituir palabras como moño a mono (no podemos utilizar la libreia unicode)
#Pasamos todo el texto a minúsculas
#Quitamos acentos y diéresis 
def estandarizacion(string):
    res=string.lower()
    res=re.sub(r"([^n\u0300-\u036f]|n(?!\u0303(?![\u0300-\u036f])))[\u0300-\u036f]+", r"\1", normalize( "NFD", res), 0, re.I)
    res=normalize( 'NFC', res)
    return res
```

```python
with open(path_lib_insultos, encoding='utf8') as f:
    lines = f.read().splitlines()
    
res_estand_lib_ins = []

# for every item in regimentNames
for i in lines:
    res_estand_lib_ins.append(estandarizacion(i))
    
res_estand_lib_ins
```

```python
print('Número de palabaras tóxicas en nuestra librería',len(res_estand_lib_ins))
```

```python
translator= Translator(from_lang="Spanish",to_lang="English")
translation = translator.translate("res_estand_lib_ins")
for sentence in res_estand_lib_ins:
    print("\n\t" +sentence+ "\t"+translator.translate(sentence)+ "\t" +str(analyser.polarity_scores(translator.translate(sentence))))
```

```python
neg_list=[]
pos_list=[]
neu_list=[]

for sentence in res_estand_lib_ins:
    neg_list.append(analyser.polarity_scores(translator.translate(sentence))['neg'] >= 0.5)
    pos_list.append(analyser.polarity_scores(translator.translate(sentence))['pos'] >= 0.5)
    neu_list.append(analyser.polarity_scores(translator.translate(sentence))['neu'] == 1.0 )

```

```python
print("Número de palabras consideradas como negativas    "+ str(sum(neg_list)))
print("Número de palabras consideradas como positivas    "+ str(sum(pos_list)))
print("Número de palabras consideradas como neutras      "+ str(sum(neu_list)))
```
